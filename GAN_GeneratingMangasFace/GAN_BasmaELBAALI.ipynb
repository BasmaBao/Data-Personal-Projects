```
import torch
import torchvision
import torch.nn as nn
from PIL import Image
import os
import torchvision.transforms as transforms
import torch.optim as optim
import random
import glob
from torch.utils.data import Dataset
from torchvision.utils import make_grid
import torchvision.utils as vutils
import matplotlib.animation as animation
from IPython.display import HTML
import numpy as np
from IPython import display
import time
```

Define custom dataset

```
class MangaFace(Dataset):
    def __init__(self, root, transform=None):
        """Initialize Pokemon."""
        self.transform = transform
        self.root = root
        self.imgs = glob.glob(self.root + '*')

    def __getitem__(self, idx):
        """Get item from Pokemon class.
        Args:
            index (int): Index
        Returns:
            img: PIL Image.
        """
        img = self.imgs[idx]
        img = Image.open(img).convert('RGB')

        if self.transform is not None:
            img = self.transform(img)

        return img

    def __len__(self):
        """Get length."""
        return len(self.imgs)

train_transform = transforms.Compose([
                                      transforms.Resize((64,64)),
                                      transforms.RandomHorizontalFlip(),
                                      transforms.ToTensor(),
                                      transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

trainset = MangaFace(root='./gan_class_images/images/', transform=train_transform)
print(len(trainset))
```

Network weights initialization

```
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        m.weight.data.normal_(0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        m.weight.data.normal_(1.0, 0.02)
        m.bias.data.fill_(0)
```

Define Generator network

```
'''
input size: batchsize * 100 * 1 * 1
output size: batchsize * 3 * 64 * 64

hint: use nn.ConvTranspose2d to upsample spatial dimension

* 1st layer:
transposed convolution
batch norm
relu

* 2nd layer
transposed convolution
batch norm
relu

* 3rd layer
transposed convolution
batch norm
relu

* 4th layer
transposed convolution
batch norm
relu

* 5th layer
transposed convolution
tanh
'''

import torch
import torch.nn as nn

class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        # TODO:

        self.net = nn.Sequential(
        # 1st layer output: bs * 512 * 4 * 4
        nn.ConvTranspose2d(100, 512, 4, 1, 0, bias=False),
        nn.BatchNorm2d(512),
        nn.ReLU(True),

        # 2nd layer output: bs * 256 * 8 * 8
        nn.ConvTranspose2d(512, 256, 4, 2, 1, bias=False),
        nn.BatchNorm2d(256),
        nn.ReLU(True),

        # 3rd layer output: bs * 128 * 16 * 16
        nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
        nn.BatchNorm2d(128),
        nn.ReLU(True),

        # 4th layer output: bs * 64 * 32 * 32
        nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
        nn.BatchNorm2d(64),
        nn.ReLU(True),

        # 5th layer output: bs * 3 * 64 * 64
        nn.ConvTranspose2d(64, 3, 4, 2, 1, bias=False),
        nn.Tanh()
        )
    def forward(self, x):

        out = self.net(x)

        return out
```

```
'''
input size: batchsize * 3 * 64 * 64
output size: batchsize * 1

* 1st layer:
convolution
relu

* 2nd layer
convolution
batch norm
relu

* 3rd layer
convolution
batch norm
relu

* 4th layer
convolution
batch norm
relu

* 5th layer
convolution
sigmoid

'''

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()

        self.net = nn.Sequential(
            # 1st layer output: bs * 64 * 32 * 32
            nn.Conv2d(3, 64, 4, 2, 1, bias=True),
            nn.ReLU(True),

            # 2nd layer output: bs * 128 * 16 * 16
            nn.Conv2d(64, 128, 4, 2, 1, bias=True),
            nn.BatchNorm2d(128),
            nn.ReLU(True),

            # 3rd layer output: bs * 256 * 8 * 8
            nn.Conv2d(128, 256, 4, 2, 1, bias=True),
            nn.BatchNorm2d(256),
            nn.ReLU(True),

            # 4th layer output: bs * 512 * 4 * 4
            nn.Conv2d(256, 512, 4, 2, 1, bias=True),
            nn.BatchNorm2d(512),
            nn.ReLU(True),

            # 5th layer output: bs * 1 * 1 * 1
            nn.Conv2d(512, 1, 4, 1, 0, bias=True),
            nn.Sigmoid()


        )
    def forward(self, x):

        out = self.net(x)

        return out.squeeze(-1).squeeze(-1)
```

Main

```
# define parameters
batch_size = 128 # 64, 32
beta1 = 0.5 # optimization function
lr = 2e-4 # learning rate
nz = 100 # input dimension
num_epochs = 10 # trainings
num_workers = 8 #

# set device
device = torch.device("cuda:0")

# load dataset
train_transform = transforms.Compose([
                                        transforms.Resize((64,64)),
                                        transforms.RandomHorizontalFlip(),
                                        transforms.ToTensor(),
                                        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

dataset = MangaFace(root='./gan_class_images/images/', transform=train_transform)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True, num_workers=num_workers, pin_memory=True)

# define G, D and initialize parameters
# attention: put it into GP
netG = Generator().to(device)
netD = Discriminator().to(device)

netG.apply(weights_init)
netD.apply(weights_init)

# define loss function
# hint: binary cross entropy
criterion = nn.BCELoss()

# Create batch of latent vectors (64 * nz * 1 * 1) that we will use to visualize
# hint: torch.randn
fixed_noise = torch.randn(64, nz, 1, 1, device=device)


# Setup Adam optimizers for both G and D
optimizer_D = optim.Adam(netD.parameters(), lr=lr, betas=(beta1, 0.999))
optimizer_G = optim.Adam(netG.parameters(), lr=lr, betas=(beta1, 0.999))


#img_list for loss
G_losses = []
D_losses = []
iters = 0

print("==> start training")
for epoch in range(num_epochs):
  # For each batch in the dataloader
  for i, data in enumerate(dataloader):
    ############################
    # Update D network
    # maximize log(D(x)) + log(1 - D(G(z)))
    ###########################
    # put data into gpu
    netD.zero_grad()
    ## Train with all-real batches
    # set D gradient 0
    real_data = data.to(device)
    batch_size = real_data.size(0)
    # set ground truth label 1
    label = torch.full((batch_size,), 1., device=device)
    # Forward pass real batch through D
    output = netD(real_data).view(-1)
    # Calculate loss on all-real batch
    errD_real = criterion(output, label)
    # Calculate gradients for D in backward pass
    errD_real.backward()

    ## Train with all-fake batch
    # Generate batch of latent vectors
    # Generate fake image batch with G
    # Classify all fake batch with D
    # set ground truth label 0
    noise = torch.randn(batch_size, nz, 1, 1, device=device)
    # Generate fake image batch with G
    fake = netG(noise)
    label.fill_(0.)
    output = netD(fake.detach()).view(-1)
    # Calculate D's loss on the all-fake batch
    # Calculate the gradients for this batch
    errD_fake = criterion(output, label)
    errD_fake.backward()
    # Add the gradients from the all-real and all-fake batches
    errD = errD_real + errD_fake
     # Update D
    optimizer_D.step()

    ############################
    # Update G network
    # maximize log(D(G(z)))
    ###########################
    # set G gradient 0
    # Since we just updated D, perform another forward pass of all-fake batch through D
    # Calculate G's loss based on this output
    # Calculate gradients for G
    # Update G
    # Save Losses for plotting later
    netG.zero_grad()
    label.fill_(1.)
    output = netD(fake).view(-1)
    errG = criterion(output, label)
    errG.backward()
    optimizer_G.step()
    # Save Losses for plotting later
    G_losses.append(errG.item())
    D_losses.append(errD.item())

    # display demo images each 50 iterations
    if iters % 50 == 0:
      display_demos(fixed_noise, epoch, iters)

    iters += 1
```

```
plt.figure(figsize=(10,5))
plt.title("Generator and Discriminator Loss During Training")
plt.plot(G_losses,label="G")
plt.plot(D_losses,label="D")
plt.xlabel("iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
